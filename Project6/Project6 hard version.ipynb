```python
import pickle
def save_pickle(model,filepath,save_name):
    '''
    Author: Huangrui Chu
    '''
    # Dump the trained decision tree classifier with Pickle
    pkl_filename = filepath+save_name+'.pkl'

    # Open the file to save as pkl file
    model_pkl = open(pkl_filename, 'wb')
    pickle.dump(model,model_pkl)

    # Close the pickle instances
    model_pkl.close()
def load_pickle(filepath,save_name):
    '''
    Author: Huangrui Chu
    '''
    classification_pkl_filename = filepath+"/"+save_name+'.pkl'
    classification_model_pkl = open(classification_pkl_filename, 'rb')
    classification_model = pickle.load(classification_model_pkl)
    #print ("Loaded HMM model: ", classification_model)
    return classification_model
```

```python
def load_all_digit_GMMHMM(filepath,filenames):
    '''
    Author: Huangrui Chu
    '''
    GMMHMMs={}
    for digit in filenames:
        print("Huangrui is loading the digit {} GMMHMM".format(digit))
        current_digit_GMMHMM=load_pickle(filepath,digit)
        GMMHMMs[str(digit)]=current_digit_GMMHMM
    return GMMHMMs
```

```python
def recognize_current_input(input_file,words,GMMHMMs):
    smallest_distance=[1000000,0]
    for word in words:
        distance=GMMHMMs[word].GMM_HMM_dtw(input_file)
        if distance<smallest_distance[0]:
            smallest_distance[0]=distance
            smallest_distance[1]=word
    recoginized_result=smallest_distance[1]
    return recoginized_result
```

```python
def getGMM_HMM_accuracy(GMMHMMs,test_data,words):
    #K-means HMM recognition
    import numpy as np
    #main
    accuracy=0
    for j in range(len(test_data)):
        current_digit=words[j%len(words)]
        recoginized_result=recognize_current_input(test_data[j],words,GMMHMMs)
        if current_digit==recoginized_result:
            accuracy+=1
        print(current_digit,'recognized as:',recoginized_result)
    print('accuracy:',accuracy/len(test_data))
```

# Continuous Speech Training

It would work for both single Gaussian and multi Gaussian.

```python
class Word:
    """docstring for word"""
    def __init__(self,word,num_of_states):
        self.word=word
        self.num_of_states=num_of_states
        self.GMMs=[]
    def add_GMM(self,GMM):
        self.GMMs.append(GMM)

class GMMinfo:
    """docstring for GMM"""
    def __init__(self,word,state):
        self.word=word
        self.state=state
        self.Gaussian_mean = []#gaussian distribution mean vector
        self.Gaussian_var = [] #gaussian distribution diagnol covarience
        self.Gaussian_weight = []#gaussian distribution weights
        self.Num_of_Gaussian = 0 # num of gaussian distribution
        self.transition_cost=np.zeros([2,1])
```

```python
class ContinuousSpeech:
    
    def __init__(self,words_list):
        '''
        Author: Huangrui Chu
        '''
        self.words_list=words_list
        self.templates={}
        self.columns={}
        
    def create_and_init_words(self,GMMHMMS):
        '''
        Author: Huangrui Chu
        '''
        #Initialize the HMMs for all digits by the models you learned for them from isolated recordings in previous assignments.
        self.GMMHMMs={}
        for word in words_list:
            init_GMM=GMMHMMS[word]
            word_gmmhmm=Word(word,init_GMM.hmm.N)
            for i in range(init_GMM.hmm.N):
                gmm=init_GMM.hmm.mix[i]
                new_gmm=GMMinfo(word,i)
                new_gmm.Gaussian_mean=gmm.Gaussian_mean
                new_gmm.Gaussian_var=gmm.Gaussian_var
                new_gmm.Gaussian_weight=gmm.Gaussian_weight
                new_gmm.Num_of_Gaussian=gmm.Num_of_Gaussian
                new_gmm.transition_cost=init_GMM.hmm.transition_cost[i+1,i+1:i+3]
                word_gmmhmm.add_GMM(new_gmm)
            self.GMMHMMs[word]=word_gmmhmm
            
    def concate_words(self,sequence):
        '''
        Author: Huangrui Chu
        '''
        self.columns[sequence]=[]
        #Translate the "0123456789" to seperate elements and then concatenat them with “silence”
        word_sequence=["silence"]
        for digit in sequence:
            word_sequence.append(digit)
        word_sequence.append("silence")
        for word in word_sequence:
            #load the word GMMs
            current_word_GMMHMM=self.GMMHMMs[word]
            for gmm in current_word_GMMHMM.GMMs:
                self.columns[sequence].append(gmm)
                
    def get_templates(self,sequence,foldername,start_instance,end_instance):
        '''
        Author: Huangrui Chu
        '''
        self.templates[sequence]=[]
        for i_th_instance in range(start_instance,end_instance):
            mfcc=getMFCC2(foldername+"/"+sequence+'_'+str(i_th_instance)+'.wav')
            self.templates[sequence].append(mfcc)
    
    def init_continous_speech_training(self,sequences,foldername,start_instance,end_instance):
        '''
        Author: Huangrui Chu
        '''
        self.sequences=sequences
        for sequence in self.sequences:
            self.get_templates(sequence,foldername,start_instance,end_instance)
            self.concate_words(sequence)
    
    def init_word_data(self):
        '''
        Author: Huangrui Chu
        '''
        self.word_data={}
        self.word_states_change={}
        for word in self.words_list:
            self.word_data[word]={}
            self.word_states_change[word]={}
            current_word_GMM=self.GMMHMMs[word]
            for state in range(current_word_GMM.num_of_states):
                #initial all the states with empty list 
                self.word_data[word][state]=[]
                self.word_states_change[word][state]=0
                #then, during the segmental kmeans, nodes would fill the list
    
    def set_pruning_threshold(self,threshold):
        self.threshold=threshold
    
    def train_continuous_speech(self):
        '''
        Input : 
        Output: just updata the gmmhmm of each digit 
        '''
        #complete set up by iteratively update the model vectors,covariance
        #and transition score
        previous_best_distance=-np.inf
        for j in range(1,100):
            current_best_distance=0 
            self.init_word_data()
            for sequence in self.sequences:
                # get the templates for current sequence
                current_templates=self.templates[sequence]
                # get the column for current sequence
                current_column=self.columns[sequence]
                #update the node state
                for k in range(len(current_templates)):
                    # deal with the k the template
                    template=current_templates[k]
                    distance=self.segmental_kmeans(current_column,template)
                    current_best_distance+=distance 
            #now, since all the word data is collected, we can now update the GMM parameters of each word
            for word in self.words_list:
                self.updata_GMMHMM(word)
                
            self.check_word_GMM_info()    
                
            difference= previous_best_distance-current_best_distance
            print("current difference")
            print(difference)
            previous_best_distance=current_best_distance
            print("updated previous best distance")
            print(previous_best_distance)
            if abs(difference)<3:
                print("Use {} iterators to updata All GMMHMM".format(j))
                break
        
        
        
    def updata_GMMHMM(self,word):
        GMMHMM=self.GMMHMMs[word]
        for state in range(GMMHMM.num_of_states):
            GMM=GMMHMM.GMMs[state]
            nodes=self.word_data[word][state]
            mix=self.GMMKmeans_WithoutEM(nodes,GMM.Num_of_Gaussian)
            #now updata the GMM using information from mix
            GMM.Gaussian_var = mix.Gaussian_var
            GMM.Gaussian_mean = mix.Gaussian_mean
            GMM.Num_of_Gaussian = mix.Num_of_Gaussian
            GMM.Gaussian_weight = mix.Gaussian_weight
            #now updata the tranistion_cost
            #first, self tranisition
            self_transition_prob=(len(nodes)-self.word_states_change[word][state])/len(nodes)
            #then, jump from state A to state B
            jump_prob=1-self_transition_prob
            GMM.transition_cost[0]=-np.log(self_transition_prob)
            GMM.transition_cost[1]=-np.log(jump_prob)
            
            
            
            
    def GMMKmeans_WithoutEM(self,nodes_for_Kmeans,num_Gaussian_distribution):
        #Huangrui Chu coded it in project3
        #initialize with mean, var and weight, with one cluster
        num_templates=len(nodes_for_Kmeans)
        means=[]
        covs=[]
        weights=[1]
        mean=np.mean(nodes_for_Kmeans,axis=0)
        cov=np.diagonal(np.cov(np.array(nodes_for_Kmeans).T),offset=0, axis1=0, axis2=1)
        means.append(mean)
        covs.append(cov)
        
        current_num_of_cluster=1
        episolom=0.04
        #initial should be 1 mean
        mix = mixInfo()
        mix.Gaussian_var = np.array(covs)
        mix.Gaussian_mean = np.array(means)
        mix.Num_of_Gaussian = current_num_of_cluster
        mix.Gaussian_weight = np.array(weights)
        stop=False
        
        while num_Gaussian_distribution>current_num_of_cluster and not stop:
            #now split
            new_means=[]
            new_covs=[]
            current_num_of_cluster=current_num_of_cluster*2
            new_clusters=[]
            for cluster in range(len(means)):
                #append newly two cluster center
                new_clusters.append([])
                new_clusters.append([])
                #get splitted mean and cov
                new_mean1=means[cluster]*(1-episolom)
                new_mean2=means[cluster]*(1+episolom)
                new_cov1=covs[cluster]*(1-episolom)
                new_cov2=covs[cluster]*(1+episolom)
                new_means.append(new_mean1)
                new_means.append(new_mean2)
                new_covs.append(new_cov1)
                new_covs.append(new_cov2)
            #now assign the templated into new clusters
            new_means=np.array(new_means)
            new_covs=np.array(new_covs)
            for node in nodes_for_Kmeans:
                d=log_gaussian(new_means,new_covs,node)
                cluster=np.argmin(d)
                new_clusters[cluster].append(node)
            #now, according to the new clustered result, we get updated weight,
            #mean and cov
            means=[]
            covs=[]
            weights=[]
            #
            print("For {} clusters, each cluster has following nodes".format(current_num_of_cluster))
            for cluster in new_clusters:
                print(len(cluster))
                if len(cluster)<2*num_Gaussian_distribution:
                    stop=True
                    print("For this state, we only have 2 Gaussian Distributions")
                mean=np.mean(cluster,axis=0)
                cov=np.cov(np.array(cluster).T)
                cov=np.diagonal(cov,offset=0, axis1=0, axis2=1)
                weight=len(cluster)/num_templates
                means.append(mean)
                covs.append(cov)
                weights.append(weight)
            #print(np.sum(weights))
            print("get {} means".format(current_num_of_cluster))
            # now, we put all the information to mix
            mix = mixInfo()
            mix.Gaussian_var = np.array(covs)
            mix.Gaussian_mean = np.array(means)
            mix.Num_of_Gaussian = current_num_of_cluster
            mix.Gaussian_weight = np.array(weights)
        return mix    
        
    def check_word_data_info(self):
        '''
        Author: Huangrui Chu
        I use this to check whether the nodes in different states of different word is reasonable
        '''
        for word in self.word_data.keys():
            current_word=self.word_data[word]
            for state in self.word_data[word].keys():
                num_nodes=len(self.word_data[word][state])
                print("for word {}, state {} has {} nodes".format(word,state,num_nodes))
                current_state_changes=self.word_states_change[word][state]
                print("for word {}, state {} has {} nodes jump to next state".format(word,state,current_state_changes))
    def check_word_GMM_info(self):
        '''
        Author: Huangrui Chu
        I use this to check whether the nodes in different states of different word is reasonable
        '''
        for word in self.words_list:
            word_GMMHMM=self.GMMHMMs[word]
            for gmm in word_GMMHMM.GMMs:
                print("for word {}, state {} has {} transition cost".format(word,gmm ,gmm.transition_cost))
                
      
    def segmental_kmeans(self,column,data):
        '''
        Author: Huangrui Chu
        '''
        #add fine
        zero39=np.zeros([data.shape[1]])
        data=np.vstack([zero39,data])
        #just initialize the trellis with inf inf ....0
        n_cols=len(data)
        n_rows=len(column)
        trellis = np.full([n_rows,n_cols], np.inf)
        trellis[0][0]=0
        
        def pruning(col):
            best=min(col)
            for i in range(len(col)):
                if col[i]>best+self.threshold:
                    col[i]= np.inf
        
        for c in range(1,n_cols):
            # pruning
            if c>=3:
                col=trellis[:,c-1]
                pruning(col)
            for r in range(n_rows):
                distance=mixture_log_gaussian(column[r],data[c])
                if r==0:
                     trellis[r][c]= trellis[r][c-1]+distance
                else:
                    #to check information
                    to_check=[trellis[r][c-1]+column[r].transition_cost[0] #self tranform
                              ,trellis[r-1][c-1]+column[r-1].transition_cost[1]]#jump from previous to current state 
                    if not min(to_check)==np.inf:
                        trellis[r][c]= min(to_check)+distance
        self.traceback(trellis,c,r,data,column)
        #Use DTW cost / frame of input speech, instead of total DTW cost, before determining threshold
        # 5 PPT  p.g 32
        min_cost=trellis[-1,-1]/len(data)
        return min_cost
        
    def traceback(self,trellis,c,r,data,column):
        #to track the nodes that jump frow word A to word B
        previous_state=column[-1].state
        while c>0 and r>0:
            to_check=[trellis[r,c-1], 
                  trellis[r-1,c-1],]
            track=np.argmin(to_check)
            if track==0:
                c-=1
            elif track==1:
                c-=1
                r-=1
            else:
                r-=1
            #segment the data
            node=data[c]
            word=column[r].word
            state=column[r].state
            self.word_data[word][state].append(node)
            if previous_state!=state:
                # in this case, current node jump from word A to word B
                self.word_states_change[word][state]+=1
                previous_state=state
        
        
        
   
```
