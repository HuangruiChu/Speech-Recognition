```python
import pickle
def save_pickle(model,filepath,save_name):
    '''
    Author: Huangrui Chu
    '''
    # Dump the trained decision tree classifier with Pickle
    pkl_filename = filepath+save_name+'.pkl'

    # Open the file to save as pkl file
    model_pkl = open(pkl_filename, 'wb')
    pickle.dump(model,model_pkl)

    # Close the pickle instances
    model_pkl.close()
def load_pickle(filepath,save_name):
    '''
    Author: Huangrui Chu
    '''
    classification_pkl_filename = filepath+"/"+save_name+'.pkl'
    classification_model_pkl = open(classification_pkl_filename, 'rb')
    classification_model = pickle.load(classification_model_pkl)
    #print ("Loaded HMM model: ", classification_model)
    return classification_model
```

```python
def load_all_digit_GMMHMM(filepath,filenames):
    '''
    Author: Huangrui Chu
    '''
    GMMHMMs={}
    for digit in filenames:
        print("Huangrui is loading the digit {} GMMHMM".format(digit))
        current_digit_GMMHMM=load_pickle(filepath,digit)
        GMMHMMs[str(digit)]=current_digit_GMMHMM
    return GMMHMMs
```

```python
def load_all_digit_GMMHMM_for_dtw(filepath,filenames):
    '''
    Author: Huangrui Chu
    '''
    GMMHMMs={}
    print("Huangrui is loading GMMHMM from folder {}".format(filepath))
    for digit in filenames:
        print("Huangrui is loading the digit {} GMMHMM".format(digit))
        current_digit_GMMHMM=load_pickle(filepath,digit)
        # make the last state only have self-transition
        current_digit_GMMHMM.hmm.transition_cost[-1,-2]=0
        GMMHMMs[str(digit)]=current_digit_GMMHMM
        
    return GMMHMMs
```

```python
def recognize_current_input(input_file,words,GMMHMMs):
    smallest_distance=[1000000,0]
    for word in words:
        distance=GMMHMMs[word].GMM_HMM_dtw(input_file)
        if distance<smallest_distance[0]:
            smallest_distance[0]=distance
            smallest_distance[1]=word
    recoginized_result=smallest_distance[1]
    return recoginized_result
```

```python
def getGMM_HMM_accuracy(GMMHMMs,test_data,words):
    #K-means HMM recognition
    import numpy as np
    #main
    accuracy=0
    for j in range(len(test_data)):
        current_digit=words[j%len(words)]
        recoginized_result=recognize_current_input(test_data[j],words,GMMHMMs)
        if current_digit==recoginized_result:
            accuracy+=1
        print(current_digit,'recognized as:',recoginized_result)
    print('accuracy:',accuracy/len(test_data))
```

# Continuous Speech Training

Here, Huangrui do implement the gamma_k. Therefore, it would work for both single Gaussian and multi Gaussian.

```python
class Word:
    """docstring for word"""
    def __init__(self,word,num_of_states):
        self.word=word
        self.num_of_states=num_of_states
        self.GMMs=[]
    def add_GMM(self,GMM):
        self.GMMs.append(GMM)

class GMMinfo:
    """docstring for GMM"""
    def __init__(self):
        self.Gaussian_mean = []#每个gaussian distribution的 mean vector
        self.Gaussian_var = [] #每个gaussian distribution的 diagnol covarience
        self.Gaussian_weight = []#每个gaussian distribution 的权重（其和为1）
        self.Num_of_Gaussian = 0 #几个gaussian distribution
        self.transition_probability=np.zeros([2,1])
```

```python
class ContinuousSpeech:
    
    def __init__(self,words_list):
        '''
        Author: Huangrui Chu
        '''
        self.words_list=words_list
        self.templates={}
        
    def create_and_init_words(self,GMMHMMS):
        '''
        Author: Huangrui Chu
        '''
        #Initialize the HMMs for all digits by the models you learned for them from isolated recordings in previous assignments.
        self.GMMHMMs={}
        for word in words_list:
            init_GMM=GMMHMMS[word]
            word_gmmhmm=Word(word,init_GMM.hmm.N)
            for i in range(init_GMM.hmm.N):
                gmm=init_GMM.hmm.mix[i]
                new_gmm=GMMinfo()
                new_gmm.Gaussian_mean=gmm.Gaussian_mean
                new_gmm.Gaussian_var=gmm.Gaussian_var
                new_gmm.Gaussian_weight=gmm.Gaussian_weight
                new_gmm.Num_of_Gaussian=gmm.Num_of_Gaussian
                new_gmm.transition_probability=np.e**(-init_GMM.hmm.transition_cost[i+1,i+1:i+3])
                word_gmmhmm.add_GMM(new_gmm)
            self.GMMHMMs[word]=word_gmmhmm
            
    def concate_words(self,sequence):
        '''
        Author: Huangrui Chu
        '''
        self.column=[]
        #Translate the "0123456789" to seperate elements and then concatenat them with “silence”
        word_sequence=["silence"]
        for digit in sequence:
            word_sequence.append(digit)
        word_sequence.append("silence")
        for word in word_sequence:
            #load the word GMMs
            current_word_GMMHMM=self.GMMHMMs[word]
            for gmm in current_word_GMMHMM.GMMs:
                self.column.append(gmm)
                
    def get_templates(self,sequence,foldername,start_instance,end_instance):
        '''
        Author: Huangrui Chu
        '''
        self.templates[sequence]=[]
        for i_th_instance in range(start_instance,end_instance):
            mfcc=getMFCC2(foldername+"/"+sequence+'_'+str(i_th_instance)+'.wav')
            self.templates[sequence].append(mfcc)
    def train_continuous_speech(self):
        '''
        Input : templates
        Output: just updata the gmmhmm of each digit 
        '''
        #complete set up by iteratively update the model vectors,covariance
        #and transition score
        previous_best_distance=-np.inf
        current_best_distance=0
        for j in range(1,100):
            #要加一个当前的sequence吧？？？？？？？？？？？？？？？？？？？？？？？？？？？？？？？
            self.alphas=[]# store the alpha_u for u in len(templates)
            self.betas=[]# store the beta_u for u in len(templates)
            self.gammas=[]# store the gamma_u for u in len(templates)
            self.gamma_ks=[]# store the gamma_u for u in len(templates)
            #update the node state
            for u in range(len(self.templates)):
                # deal with the k the template
                self.Baum_Welch(self.templates[u])
        #now, we have all the gamma_u and gamma_k_u
        #let me updata HMM Parameters  page 174
        
        #先更新 mu~k,s
        S=len(self.column) # This is the S
        for s in range(S):
            current_GMM=self.column[s]
            for k in range(current_GMM.Num_of_Gaussian):
                gamma_ksut=[]
                gamma_ksut_X_ut=[]
                for u in range(len(self.templates)):
                    gamma_ksut_X_ut.append(self.gamma_ks[s,:,k]*self.templates[u],axis=1)
                    gamma_kust.append(np.sum(self.gamma_ks[s,:,k],axis=1))
                    
            
            
                
    def gaussian(self,mu,squared_sigma,input_vector):
        #Author: Huangrui Chu, 
        #Calculate the probability, we only return a number!!!!
        #为了方便 我这边的一个numpy 推广就先应用到mu 上吧 毕竟我后面是一帧一帧的去分析
        #print(type(squared_sigma))
        d=input_vector.shape[0]# which means 39
        part0=np.prod(squared_sigma,axis=1)# Huangrui use this to give the product of squared_sigma 1,2,...39
        part1=np.sqrt((2*np.pi)**d *part0)
        front=1/part1
        part2=0.5*np.sum((mu-input_vector)**2/squared_sigma,axis=1)
        expo=np.exp(-part2)
        p=front*expo
        #p=np.exp(log_gaussian(mu,squared_sigma,x))
        return p
    
    def log_gaussian(self,mu,squared_sigma,input_vector):
        #Author: Huangrui Chu, 
        #Calculate the cost using log gaussian
        part1=0.5*np.sum(np.log((2*np.pi)*(squared_sigma)),axis=1)
        part2=0.5*np.sum(np.square((input_vector-mu))/squared_sigma,axis=1)
        cost= part1+part2
        return cost
    
    def probability(self,X,GMM):
        '''
        Input : 
        X: is x_t, which means the observation data at time t
        GMM: is the GMM of state s
        Output: the probability density of x_t is generated by current GMM
        '''
        weight=GMM.Gaussian_weight
        mu = GMM.Gaussian_mean
        squared_sigma = GMM.Gaussian_var
        probability=self.gaussian(mu,squared_sigma,X)
        weighted_probability=np.sum(weight*probability)
        return weighted_probability
    
    def log_probability(self,X,GMM):
        '''
        Input : 
        X: is x_t, which means the observation data at time t
        GMM: is the GMM of state s
        Output: the log probability density of x_t is generated by current GMM
        the reason Huangrui use log probability is that when the cov is small, probability is tooo large
        '''
        weight=GMM.Gaussian_weight
        mu = GMM.Gaussian_mean
        squared_sigma = GMM.Gaussian_var
        probability=self.log_gaussian(mu,squared_sigma,X)
        weighted_cost=np.sum(weight*probability)
        return weighted_cost
    
    
    def Baum_Welch(self,data):
        #Huangrui Chu use GMM
        #Author: Huangrui Chu
        T=len(data) # This is the T
        S=len(self.column) # This is the S
        alpha=np.zeros([S,T])
        beta=np.zeros([S,T])
        gamma=np.zeros([S,T])
        gamma_k=np.zeros([S,T,5])
        pi=np.zeros(S)
        pi[0]=1
        #some denotion introduction:
        #alpha (s,t) is the total log_probability of ALL state sequences that end at state s at time t, and all observations until x 
        #column[s].transition_log_probability[0] is P(s|s)
        #column[s-1].transition_log_probability[1] is P(s|s') where s' is the previous state
        #self.log_probability(data[t],self.column[s]) is the P(X_t|s), the generating log_probability
        #alpha (s,1) is simply the log_probability of being in state s at t=1 and generating obervation x1 from s
        
        
        #The Forward Algorithm
        #Initialize all alpha terms at t=1:
        for s in range(S):
            alpha[s,0]=pi[s]*self.log_probability(data[0],self.column[s])
        
    
        # Recursive estimate alphas for all subsequent time steps
        for t in range(1,T):
            for s in range(0,S):
                if s==0:
                    #if it is the initial state，there is no s'
                    alpha[s,t]=alpha[s,t-1]*self.column[s].transition_probability[0]*self.log_probability(data[t],self.column[s])
                else:
                    alpha[s,t]=(alpha[s,t-1]*self.column[s].transition_probability[0]+
                                alpha[s,t-1]*self.column[s-1].transition_probability[1])*self.log_probability(data[t],self.column[s])
        #The Backward Algorithm   
        
        #Note: Beta for any time t does not include the contribution of the observation at that time!!!!! page 163
        #Initialize all beta terms at t=T:
        beta[:,T-1]=1
        #Recursively estimate betas for all prior time steps
        for t in range(T-2,-1,-1):
            for s in range(0,S):
                if s==S-1:
                    #which means it is the end state
                    beta[s,t]=beta[s,t+1]*self.column[s].transition_probability[0]*self.log_probability(data[t+1],self.column[s])
                else:
                    #up, the later one state; down, the current state 
                    up=beta[s+1,t+1]*self.column[s].transition_probability[1]*self.log_probability(data[t+1],self.column[s+1])
                    print("log_probability is of later state{}".format(self.log_probability(data[t+1],self.column[s+1])))
                    down=beta[s,t+1]*self.column[s].transition_probability[0]*self.log_probability(data[t+1],self.column[s])
                    print("log_probability is {}".format(self.log_probability(data[t+1],self.column[s+1])))
                    beta[s,t]=up+down
        #gamma
        #The gamma at any state at any time is obtained by normalizing the product of alphas and betas to sum to 1.0 over the corresponding column of the trellis
        for t in range(1,T):
            for s in range(0,S):
                if s==0:
                    #if it is the initial state，there is no s'
                    gamma[s,t]=1
                    # now for K 在这边加k吗？
                    for k in range (self.column[s].Num_of_Gaussian):
                        up=self.log_gaussian(self.column[s].Gaussian_mean[k],self.column[s].Gaussian_var[k],data[t])
                        down=self.log_probability(data[t+1],self.column[s+1])
                        gamma_k[s,t,k]=gamma[s,t]*(up/down)
                else:
                    gamma[s,t]=alpha[s,t]*beta[s,t]/(alpha[s,t]*beta[s,t]+alpha[s-1,t]*beta[s-1,t])
                    # now for K 在这边加k吗？
                    for k in range (self.column[s].Num_of_Gaussian):
                        up=self.log_gaussian(self.column[s].Gaussian_mean[k],self.column[s].Gaussian_var[k],data[t])
                        down=self.log_probability(data[t+1],self.column[s+1])
                        gamma_k[s,t,k]=gamma[s,t]*(up/down)

        self.alphas.append(alpha)
        self.betas.append(beta)
        self.gammas.append(gamma)
        self.gamma_ks.append(gamma_k)
        
 

        
```
